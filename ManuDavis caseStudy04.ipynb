# Case Study 04 - Machine Learning to predict public sentiment from text data

```python
import numpy as np 
import pandas as pd
```

```python
#loading data
data = pd.read_csv("D:\Machine Learning\Case study\Case study 04\judge-1377884607_tweet_product_company.csv", encoding='latin-1')
```

```python
data.head()
```

```python
pd.set_option('display.max_colwidth', None)
tweets = data[['tweet_text','is_there_an_emotion_directed_at_a_brand_or_product']]
tweets.columns = ["text","label"]
tweets.head()
tweets['text'] = tweets['text'].astype(str)
```

```python
tweets.label.unique()
```

```python
tweets['label'].isnull().values.any()
```

```python
tweets['label']=tweets['label'].map({'Negative emotion':0,'Positive emotion':1,'No emotion toward brand or product':2,"I can't tell":3})
```

```python
tweets.head()
```

```python
tweets.shape
```

```python
from keras.preprocessing import text
tokenizer = text.Tokenizer() 
tokenizer.fit_on_texts(list(tweets['text']))
tokenized_texts = tokenizer.texts_to_sequences(tweets['text'])
```

```python
tokenized_texts[3]
```

```python
len(tokenized_texts[3])
```

```python
#padding
from keras_preprocessing.sequence import pad_sequences
X = pad_sequences(tokenized_texts, maxlen=100)
```

```python
X[3]
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, tweets['label'], test_size=0.2)
```

```python
from keras import Sequential, layers
from keras.layers import Dense, LSTM, Embedding, SimpleRNN, Dropout
```

# simpleRNN

```python
model = Sequential()

model.add(Embedding(input_dim = len(tokenizer.word_index)+1, output_dim = 128))
model.add(SimpleRNN(10))
model.add(Dense(50, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))  
```

```python
model.compile(loss='binary_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=10, validation_split=0.1)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[0]
```

```python
y_pred[y_pred>0.5] = 1
y_pred[y_pred<=0.5] = 0
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

# LSTM

```python
model = Sequential()

model.add(Embedding(input_dim = len(tokenizer.word_index)+1, output_dim = 128,input_length=100))
model.add(LSTM(10))
model.add(Dropout(0.5))
model.add(Dense(50, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))  
```

```python
model.compile(loss='binary_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, validation_split=0.1)
```

```python
y_pred = model.predict(X_test)
y_pred[y_pred>0.5] = 1
y_pred[y_pred<=0.5] = 0
accuracy_score(y_pred,y_test)
```
